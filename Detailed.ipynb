**How to create a new repository (repo means folder)**

- `git init` This will create a new folder with a .git folder inside it. This is where all the information about your repo is stored.

Now, you can add files to your repo. You can do this by creating a file in the folder and then adding it to the repo.

Now, let's add a file to git repo. First, create a file in the folder. We can create the file using windows command `echo "Hello World" > hello.txt` . This will create a file named `hello.txt` with the content `Hello World` in it.


Now because we already ordered git to track the files in the folder, we can check the status of the files using `git status` command. This will show the status of the files in the folder. The file we just created will be shown as untracked. This means that git is not tracking the changes in the file. To track the changes in the file, we need to add the file to the staging area. 

To add the file to the staging area, we use `git add hello.txt` command. This will add the file to the staging area. Now, if we check the status of the files using `git status` command, we can see that the file is added to the staging area.

Now, we need to commit the changes to the repo. To do this, we use `git commit -m "Initial commit"` command. This will commit the changes to the repo. Now, if we check the status of the files using `git status` command, we can see that the file is committed to the repo.

We can also check the log of the commits using `git log` command. This will show the log of the commits.


Now, go and change the content of the file and stage them . If we want to unstage the changes, we can use `git reset hello.txt` command. This will unstage the changes. If we want to unstage all the changes, we can use `git reset` command. This will unstage all the changes.

Now, if we want to revert the changes, we can use `git checkout hello.txt` command. This will revert the changes. If we want to revert all the changes, we can use `git checkout .` command. This will revert all the changes.

We can stash the changes using `git stash` command. This will stash the changes. We can check the stash using `git stash list` command. This will show the list of stashes. We can apply the stash using `git stash apply` command. This will apply the stash. We can also delete the stash using `git stash drop` command. This will delete the stash.

Now, if we want to restore repo to a previous commit, we can use `git reset <commit-id>` command. This will restore the repo to the previous commit. But the files will be in the staging area. To unstage the files, we can use `git reset` command. This will unstage the files and restore the repo to the previous commit.Or we can send the staged changes to stash using `git stash` command. 

These are some of the basic commands of git. There are many more commands in git. But these are the most used commands.

Why need Github when we have git?
- Github is a cloud based service that hosts git repositories. It provides a lot of  features like collaboration, issue tracking, etc. But git is a command line tool that is used to manage the changes in the files. It is a version control system. It is used locally. 

First, we need to create a github account. Then, we need to create a repository in github. Then, we need to add the remote repository to our local repository. To do this, we use `git remote add origin <url>` command. This will add the remote repository to our local repository. Now, we can push the changes to the remote repository using `git push origin master` command. This will push the changes to the remote repository. We can also pull the changes from the remote repository using `git pull origin master` command. This will pull the changes from the remote repository.

Now, we can clone the remote repository to our local repository using `git clone <url>` command. This will clone the remote repository to our local repository. We can also fork the remote repository to our github account. This will create a copy of the remote repository in our github account.

**Branching**
We can create a branch using `git branch <branch-name>` command. This will create a branch. We can switch to the branch using `git checkout <branch-name>` command. This will switch to the branch. We can also create a branch and switch to it using `git checkout -b <branch-name>` command. This will create a branch and switch to it. We can also delete a branch using `git branch -d <branch-name>` command. This will delete the branch.
